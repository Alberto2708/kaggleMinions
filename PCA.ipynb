# PCA 
### El propósito es ver cuanto podríamos conseguir y como podríamos distribuir los datos

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python

base = pd.read_csv("Datasets/Base.csv")
variantI = pd.read_csv("Datasets/Variant I.csv")
variantII = pd.read_csv("Datasets/Variant II.csv")
variantIII = pd.read_csv("Datasets/Variant III.csv")
variantIV = pd.read_csv("Datasets/Variant IV.csv")
variantV = pd.read_csv("Datasets/Variant V.csv")
```

```python
dataset = base
```

```python
target = dataset['fraud_bool']
features = dataset.drop(columns=['fraud_bool'])

categorical_columns = features.select_dtypes(include=['object']).columns.tolist()
numerical_columns = features.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_dataset = features.drop(columns=categorical_columns)

encoder = OneHotEncoder(sparse_output=False, dtype=int)
one_hot_encoded = encoder.fit_transform(features[categorical_columns])

one_hot_encoded_df = pd.DataFrame(
    one_hot_encoded, 
    columns=encoder.get_feature_names_out(categorical_columns),
    index=features.index
)

dataset_encoded = pd.concat([numerical_dataset, one_hot_encoded_df, target], axis=1)
dataset_encoded.info()
```

```python
#Si quieres cambiar a probar solo con las variables numericas, cambia la linea de scaled_data 
scaler = StandardScaler()
#scaled_data = scaler.fit_transform(numerical_dataset)
scaled_data = scaler.fit_transform(dataset_encoded)
pca = PCA(n_components=30)
pca.fit(scaled_data)
```

```python
print("Componentess con maxima varianza:")
for i,c in enumerate(pca.components_):
  print(f"Componente {i} = {c}")

print("\nMagnitud de varianza explicado por cada uno de los componentes seleccionados:")
for i,m in enumerate(pca.explained_variance_):
  print(f"Magnitud del componente {i} = {m}")
```

```python
for i, n in enumerate(np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)):
  print(f"componente {i}: {n}%")
```

# Diferentes intentos de modelos de clasificación

```python
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42, stratify=target
)
```

```python
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_train_encoded = encoder.fit_transform(X_train[categorical_columns])
X_test_encoded = encoder.transform(X_test[categorical_columns])
```

```python
# 2. Combine with numerical features
X_train_processed = np.hstack([X_train[numerical_columns].values, X_train_encoded])
X_test_processed = np.hstack([X_test[numerical_columns].values, X_test_encoded])
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_processed)
X_test_scaled = scaler.transform(X_test_processed)
```

```python
pca = PCA(n_components=30)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
model = svm.SVC(class_weight='balanced')  # Handle class imbalance
model.fit(X_train_pca, y_train)

# Use proper evaluation
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

y_pred = model.predict(X_test_pca)
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print(f"ROC-AUC: {roc_auc_score(y_test, y_pred)}")
```
